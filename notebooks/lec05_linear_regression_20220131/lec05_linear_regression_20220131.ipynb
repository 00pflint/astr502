# Quick Introduction

Linear Neural Network are dealing with primarily two different problems, i.e. linear regression and classification regression.  <br>
Linear regression is used to analyze data with **continuous** target variable $y$, and hope ML to provide **prediction** to arbitary input, in textbook Section 3.1-3.3. <br>
Classification, on the other hand, is a technique to deal with data with **discrete** (or **categorized**) target variable $y$, and will be discussed by Chia-Lin in next lecture, and in textbook Section 3.4-3.7.

## Naming Convention

To simplify the content and sync everyone with the textbook reading, I am going to follow the naming given by the book and we are using them throughout the rest of the content. <br>
1. $\mathbf{x}= (x_1, x_2, ..., x_d)$ : **independent variables** or **feature vector**  <br>  $y$ : **response variable** or **target label** or **dependent variable** 
2. Superscript ${(i)}$ refers to a particular training example $i$. ($\mathbf{x}^{(i)}$, $y^{(i)}$) is a paired feature vector and target label for training example $i$. <br>
A dataset with $n$ entries can be expressed as 
$\{(\mathbf{x}^{(1)},y^{(1)}),(\mathbf{x}^{(2)},y^{(2)}),......,(\mathbf{x}^{(i)},y^{(i)}),......,(\mathbf{x}^{(n)},y^{(n)})\}$
3. Subscript $j$ refers to the $j$-th component of the feature vector $\mathbf{x}$. <br>
   e.g. $x^{(i)}_2$ refers to the 2nd component of training example $\mathbf{x}^{(i)}$
4. Letter with hat, $\hat{y}$ refers to estimated value, as opposed to original value $y$.
5. Bold symbols of $x$, known as vector or matrix version, have slightly different meaning in different expression. $\mathbf{x}$ refers to a row-wise or column-wise vector, while $\mathbf{X}$ refers to the whole sample set.
6. $\mathbf{w}$ is the weight vector (of the linear model) and $b$ is bias or y-intercept.
7. Majority of times without stating, $n$ refers to number of data entries or size of row in matrix, $d$ refers to dimensionality of features or size of column in matrix.
8. In this chapter, unless specified, $\mathbf{b}$ is always a scalar multiple a 1-row/column vector, i.e. $\mathbf{b}= b \begin{pmatrix}1 & 1 & ...\end{pmatrix}$, a broadcasting scalar.
9. A dataset can be separated into training set for linear regression model training, and testing set for validation or prediction.

# General Concept

## Loss function

Given a training set of features ($\mathbf{X}$) and labels ($\mathbf{y}$), we are trying to characterize the relationships between them by implementing a linear relationship as following, 
\begin{align*}
\mathbf{\hat{y}} = \mathbf{Xw} + \mathbf{b} 
\end{align*}
where $\mathbf{\hat{y}}$ is the estimated labels. <br>

To understand how good we model the relationship, we require some sort of metric to evaluate the goodness. First we look at how well our model can constrain the linear relationship. <br>
Observation always accompany with noise, given dataset $\{(\mathbf{x},y)\}$ (one row data), we have 
\begin{align*}
\epsilon = y - \hat{y}
\end{align*}
Assume the noise is **normally distributed**, $\epsilon \sim \mathcal{N} (0,\sigma^2)$, therefore how likely are we going to observe $y$ given $\mathbf{x}$ is characterized as the conditional probability of $P(y |\mathbf{x} )$, which should follow the underlying error's normal distribution,
\begin{align}
P(y |\mathbf{x} ) = \frac{1}{\sqrt{2 \pi \sigma^2}} \mathrm{exp} \left[\frac{-(y-\hat{y})^2}{2\sigma^2}\right]
\end{align}
This is called **likelihood** function of $\{(\mathbf{x},y)\}$ with linear modelling. <br>

In most of the cases errors can be safely assume to be **independent** between each sample set (i.e. $y^{(1)}$ error and $y^{(2)}$ error are uncorrelated) and identically distributed (drawn from the same error population, so **same $\sigma$**), for the entire training set with $\mathbf{X}$ and $\mathbf{y}$,
\begin{align}
P(\mathbf{y}|\mathbf{X}) &= \prod_{i=1}^{n} P(y^{(i)}|\mathbf{x}^{(i)}) \nonumber \\
&= \prod_{i=1}^{n} \frac{1}{\sqrt{2 \pi \sigma^2}} \mathrm{exp} \left[\frac{-(y^{(i)}-\hat{y}^{(i)})^2}{2\sigma^2}\right]
\end{align}
The best model parameters $\mathbf{w,b}$ are those maximize the whole training set likelihood function $P(\mathbf{y}|\mathbf{X})$. It would be mathematically easier to maximize the $\log$ of $P(\mathbf{y}|\mathbf{X})$ since that does not change the result either. <br>
\begin{align}
\log P(y |\mathbf{x} ) &= \log \left( \frac{1}{\sqrt{2 \pi \sigma^2}} \mathrm{exp} \left[\frac{-(y-\hat{y})^2}{2\sigma^2}\right] \right) \nonumber \\
&= \sum_{i=1}^{n} \frac{-1}{2} \log(2\pi\sigma^2) - \sum_{i=1}^{n} \frac{1}{2 \sigma^2} (y^{(i)}-\hat{y}^{(i)})^2 \nonumber \\
&= -\frac{n}{2} \log \left(2 \pi \sigma^2\right) - \frac{1}{\sigma^2} \frac{1}{2} \sum_{i=1}^{n} \left(y^{(i)}-\hat{y}^{(i)}\right)^2
\end{align}

Assume $\sigma$ is constant, from above equation we know that maximizing the entire training set log-likelihood is equivalent to minimizing the overall squared error between estimated and true labels. <br>
Therefore it is natural to define a **loss function** in linear regression, as 
\begin{align}
\ell^i = \frac{1}{2}\left(y^{(i)}-\hat{y}^{(i)}\right)^2 
\end{align}
for single sample set. For entire set,
\begin{align}
L(\mathbf{w,b})=\frac{1}{n} \sum\limits_{i=1}^{n} \ell^{i} = \frac{1}{n} \sum\limits_{i=1}^{n} \frac{1}{2} \left(\mathbf{y}^{(i)} - \mathbf{w}^{\top}\mathbf{x}^{(i)} - \mathbf{b}  \right)^2
\end{align}
and the goal is find 
$$(\mathbf{w^{\ast},b^{\ast}}) = \operatorname*{argmin}_{\mathbf{w, b}} \;L(\mathbf{w,b)} $$
**Conclusion in Chapter 3.1: Minimize the mean squared error is equivalent to maximize likelihood estimation of a linear model under the assumption of additive Gaussian noise.**


Comment from Tim + Hung-Jin: Knowing the $\sigma$ value would not affect the best-fit parameter estimations on ${\mathbf w^*}$ and $b^*$. But it helps in constraining the uncertainty of our model parameters (i.e. the error bars of ${\mathbf w^*}, b^*$), and thus knowing how confident we can believe our best-fit model given our data uncertainty.

##  Optimization: Gradient Descent

In linear regression one of the most common method to find $\mathbf{w},b$ is to miniziming the loss function through gradient descent. Simply speaking, each time loss function $L$ is computed, we compute gradient of loss function with respect to $\mathbf{w},b$, and update $\mathbf{w},b$ by subctrating the gradient off.<br>
Besides that, we also need to control how far each update goes. Sometimes the gradient may be too extreme and the optimum value just passes off, so we need to adjust **learning rate** $\eta$, which controls the step for update process.
\begin{align}
(\mathbf{w}, b) &\leftarrow (\mathbf{w}, b) - \eta \frac{\partial \; L(\mathbf{w,b})}{\partial (\mathbf{w,b})} \nonumber \\
w_j &\leftarrow w_j - \eta \frac{1}{n} \sum_{i=1}^{n} \frac{\partial \; \ell^{(i)} (\mathbf{x^{(i)}},y^{(i)},\mathbf{w})}{\partial w_j}
\end{align}
where n is the training set size (I abosrb $b$ into $w$ in element form). This is called full batch gradient descent. Everytime when whole training set is completely explored once (**1 epoch**) the weights and bias parameter update according to gradient of loss function with respect to whole training set.

The **full batch gradient descent** could be slow when the training set has monster size. One would expect another extreme situation: **stochastic gradient descent**. <br>
That is, for every single sample set ($\mathbf{x}^{(i)},y^{(i)}$), weights and bias update accordingly.
\begin{align}
w_j &\leftarrow w_j - \eta \frac{\partial \; \ell^{(i)} (\mathbf{x^{(i)}},y^{(i)},\mathbf{w})}{\partial w_j}
\end{align}
Because each sample set may have different optimum value, weights and bias would have shown some seesaw behaviour or non-converging issue or give unstable solution.

A standard approach is **minibatch stochastic gradient descent**. Simply speaking, we choose a batch size (usually in power of 2, based on the memory on GPU that can be used for parallelization to speed up training process) and randomly select samples from training set to form a minibatch. We use this minibatch to compute overall loss function, and update model parameters. We repeat this process (**1 iteration**), disregard those been selected in previous routines, with roughly $\displaystyle\frac{\text {training set size}}{\text{minibatch size}}$ times, so we make use of all samples information in training set (**1 epoch**). <br>
\begin{align}
w_j &\leftarrow w_j - \eta \frac{1}{|\mathcal{B}|} \sum_{i \in \mathcal{B}} \frac{\partial \; \ell^{(i)} (\mathbf{x^{(i)}},y^{(i)},\mathbf{w})}{\partial w_j}
\end{align}

## Synthetic Data Generation

```python
def synthetic_data(w, b, num_examples):  #@save
    """Generate y = Xw + b + noise."""
    X = torch.normal(0, 1, (num_examples, len(w)))
    y = torch.matmul(X, w) + b
    y += torch.normal(0, 0.01, y.shape)
    return X, y.reshape((-1, 1))

true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = synthetic_data(true_w, true_b, 1000)
```

```python
# Take a look on dataset distribution.
fig, ax = d2l.plt.subplots(1,2,figsize=(8,5))
# The semicolon is for displaying the plot only
ax[0].scatter(features[:, (0)].detach().numpy(), labels.detach().numpy(), 1);
ax[1].scatter(features[:, (1)].detach().numpy(), labels.detach().numpy(), 1);
ax[0].set_xlabel(r'$x_1$');ax[1].set_xlabel(r'$x_2$');ax[0].set_ylabel(r'$y$');ax[1].set_ylabel(r'$y$');
```

## Use data_iter to read dataset and output minibatch

```python
# ### Method 1 
# def data_iter(batch_size, features, labels):
#     num_examples = len(features)
#     indices = list(range(num_examples))
#     # The examples are read at random, in no particular order
#     random.shuffle(indices)
#     for i in range(0, num_examples, batch_size):
#         batch_indices = torch.tensor(
#             indices[i: min(i + batch_size, num_examples)])
#         yield features[batch_indices], labels[batch_indices]

# print (f'features shape:{features.shape}, labels shape:{labels.shape}')

# # Determine the number of minibatch, or the number of partition you want to cut the data into
# batch_size = 10
## You can check one minibatch feature and label set by uncommenting next line.
# for X, y in data_iter(batch_size, features, labels):
#     print(X, '\n', y)
#     break
    
### Method 2 
# Use API to read the data, and it becomes an iterator to perform shuffling or randomize data order 
# by it own everytime we call it for training.
def load_array(data_arrays, batch_size, is_train=True):  #@save
    """Construct a PyTorch data iterator."""
    dataset = data.TensorDataset(*data_arrays)
    return data.DataLoader(dataset, batch_size, shuffle=is_train)

batch_size = 10
data_iter = load_array((features, labels), batch_size)
# You can check one minibatch feature and label set by uncommenting next line.
# next(iter(data_iter))
```

## Define Neural Network Model

```python
# Multiple possible NN models (or called layers) available at https://pytorch.org/docs/stable/nn.html.
# Here we use the linear layer.
# (2,1) defines the number of input feature dimension, 1 is the output feature dimension.
net = nn.Sequential(nn.Linear(2, 1))
```

## Initialize Model Parameters $w$ and $b$

```python
print (f'Initial weight guess: {net[0].weight.data.normal_(0, 0.01)}')
print (f'Initial bias guess: {net[0].bias.data.normal_(0, 0.01)}')
```

## Define Loss Function

```python
# There are multiple available loss functions defined in pytorch, refer to 
# https://pytorch.org/docs/stable/nn.html#loss-functions
# for more detail.
# We use mean squared error in this work. 
loss = nn.MSELoss()
```

## Define Optimization Algorithm

```python
trainer = torch.optim.SGD(net.parameters(), lr=0.03)
```

## Hyperparameters

There are two parameters here we have to manually decide and has nothing to do with model parameters: **batch sizes** and **learning rate**, so called hyperparameters. The values can be chosen arbitarily, and it affects how efficient the training process and sometimes may affect the outcome. The setting requires some prior knowledge or experiences or trial and error to come out with good tuning value. Here we set batch_size = 10, and learning rate = 0.03.

## Trainning Process

Quote from the book Chap 3.3.7: <br>
To refresh your memory: for some number of epochs,
we will make a complete pass over the dataset (`train_data`),
iteratively grabbing one minibatch of inputs
and the corresponding ground-truth labels.
For each minibatch, we go through the following ritual:

* Generate predictions by calling `net(X)` and calculate the loss `l` (the forward propagation).
* Calculate gradients by running the backpropagation.
* Update the model parameters by invoking our optimizer.

For good measure, we compute the loss after each epoch and print it to monitor progress.


```python
iteration_loss = []
num_epochs = 3
iteration_count = 0
for epoch in range(num_epochs):
    for X, y in data_iter:
        l = loss(net(X) ,y)
        iteration_count += 1
        iteration_loss.append(l.item())
        trainer.zero_grad()
        l.backward()
        trainer.step()
    l = loss(net(features), labels)
    print (f'After {iteration_count} iteration we enter next epoch.')
    print(f'epoch {epoch + 1}, loss {l:f}')
    
```

```python
d2l.use_svg_display()
fig, ax = d2l.plt.subplots(1,1,figsize=(8,6))
d2l.set_axes(axes=ax,xlabel='iteration',ylabel='loss',xlim=((0,iteration_count+1)),ylim=((min(iteration_loss),max(iteration_loss))),xscale='linear',yscale='log',legend=None)
d2l.plt.plot(range(1,iteration_count+1),iteration_loss)
```

```python
w = net[0].weight.data
print(f'error in estimating w: {(true_w - w.reshape(true_w.shape)).tolist()}, around {abs(sum(((true_w - w.reshape(true_w.shape))*100./true_w).tolist())/len(true_w)):.2e} %.')
b = net[0].bias.data
print(f'error in estimating b: {(true_b - b).item()}, around {((true_b - b)*100./true_b).item():.2e} %.')
```

# Compact Way

```python
batch_size_list = [1,8,32]
learning_rate_list = [0.01,0.03,0.1]
iteration_loss_list = []
computation_time_list = []

for batch_size in batch_size_list:
    for learning_rate in learning_rate_list:
        data_iter = load_array((features, labels), batch_size)
        net = nn.Sequential(nn.Linear(2, 1))
        loss = nn.MSELoss()
        trainer = torch.optim.SGD(net.parameters(), lr=learning_rate)
        num_epochs = 3
        iteration_count = 0
        iteration_loss = []
        computation_time = []
        start = time.time()
        for epoch in range(num_epochs):
            for X, y in data_iter:
                l = loss(net(X) ,y)
                iteration_count += 1
                iteration_loss.append(l.item())
                trainer.zero_grad()
                l.backward()
                trainer.step()
                computation_time.append(time.time()-start)
            l = loss(net(features), labels)
        end = time.time()
        computation_time_list.append(computation_time)
        iteration_loss_list.append(iteration_loss)
```

```python
d2l.use_svg_display()
fig, ax = d2l.plt.subplots(1,2,figsize=(12,6))
d2l.set_axes(axes=ax[0],xlabel='iteration',ylabel='loss',xlim=((0,301)),ylim=((1.e-6,1.e2)),xscale='linear',yscale='log',legend=None)
ax[0].plot(range(1,len(iteration_loss_list[1])+1),iteration_loss_list[1],label='batch_size=1')
ax[0].plot(range(1,len(iteration_loss_list[4])+1),iteration_loss_list[4],label='batch_size=8')
ax[0].plot(range(1,len(iteration_loss_list[7])+1),iteration_loss_list[7],label='batch_size=32')
ax[0].legend();

d2l.set_axes(axes=ax[1],xlabel='iteration',ylabel='loss',xlim=((0,301)),ylim=((1.e-6,1.e2)),xscale='linear',yscale='log',legend=None)
ax[1].plot(range(1,len(iteration_loss_list[3])+1),iteration_loss_list[3],label='lr=0.01')
ax[1].plot(range(1,len(iteration_loss_list[4])+1),iteration_loss_list[4],label='lr=0.03')
ax[1].plot(range(1,len(iteration_loss_list[5])+1),iteration_loss_list[5],label='lr=0.1')
ax[1].legend();
```
